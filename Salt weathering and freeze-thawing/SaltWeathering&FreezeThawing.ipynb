```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#Salt weathering
dfRH = pd.read_csv("RH Trondheim.csv")
RH_array = dfRH.to_numpy()

cycles = 0
for i in range(len(RH_array) - 1):
    if((RH_array[i] < 75 and RH_array[i+1] > 75) or (RH_array[i] > 75 and RH_array[i+1] < 75)):
        cycles += 1

print(cycles)
        
```

```python
#Freezing-thawing
dfT = pd.read_csv("T Trondheim.csv", delimiter=";")
T_array = dfT.to_numpy()
float_array = np.array([[float(value.replace(',', '.')) for value in row] for row in T_array])

"""cycles = 0
for i in range(len(float_array) - 1):
    if((float_array[i] > 0 and float_array[i+1] < 0) or (float_array[i] < 0 and float_array[i+1] > 0)): 
        cycles += 1


print(cycles)"""

cycles = 0
i = 0
while i < len(float_array):
    if float_array[i] <= -3:
        j = 1
        while i + j < len(float_array) and float_array[i + j] < 1:
            j += 1
        if i + j < len(float_array) and float_array[i + j] >= 1:
            cycles += 1
        i = i + j
    elif float_array[i] >= 1:
        k = 1
        while i + k < len(float_array) and float_array[i + k] > -3:
            k += 1
        if i + k < len(float_array) and float_array[i + k] <= -3:
            cycles += 1
        i = i + k
    else:
        i += 1
    
print(cycles)
```
